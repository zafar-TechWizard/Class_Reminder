```python
semester = '5'
stream = ' CSE '
section = 'E1'

batch = semester + stream +section
```

```python
def fil_nan(x):
    return 'Filled' if pd.isna(x) else x


def load_and_preprocess(file, batch):
    TimeTable = pd.read_excel(file)
    # Get the data
    df = TimeTable[['Day', 'Timmings', batch]]
    df = df.copy()
    # Remove gap between days
    df = df.dropna(subset=['Timmings'])

    # Fixing indices
    df.reset_index(drop=True, inplace=True)

    days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

    for d in days:
        get_index = df[df['Day']== d].index[0]
        df.loc[get_index : get_index+7, 'Day'] = d

    df['5 CSE E1'] = df['5 CSE E1'].apply(fil_nan)
    
    for i in df.loc[df['5 CSE E1'] == 'Filled'].index:
        print(df['5 CSE E1'][i] ,": ", df['5 CSE E1'][i-1], "\n")
        df.at[i, '5 CSE E1'] = df['5 CSE E1'][i-1]


    
    df.to_csv("data.csv")

    return
```

```python
load_and_preprocess('Timetable.xlsx', batch=batch)
```

```python
def AmPm(time):
    hr, mnt = map(int, time.split(':'))

    if hr ==8 or hr == 9 or hr == 10 or hr == 11:
        return f"{hr}:{mnt:02d} AM"
    
    elif hr == 12:
        return f"{hr}:{mnt:02d} PM"
        
    else:
        return f"{hr}:{mnt:02d} PM"
```
